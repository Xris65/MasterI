
<p>A présent, nous allons illustrer la puissance de le bibliothèque <code>numpy</code>. Pour cela nous allons d'abord définir notre propre fonction <code>prod_matrices</code> calculant le produit en utilisant des boucles imbriquées :</p>


```python

def prod_matrices(M1, M2):
    M = np.zeros((M1.shape[0], M2.shape[1]))
    for i in range(M1.shape[0]):
        for j in range(M2.shape[1]):
            for k in range(M1.shape[1]):
                M[i, j] += M1[i, k] * M2[k,j] 
    return M

```


<p>Test de la fonction :</p>


```python

M1 = np.array([[2, 1, 1], [4, 5, 7]])
M2 = np.array([[3, 2], [2, 1], [5, 4]])
prod_matrices(M1, M2)

```


<p>Avec un peu plus d'éléments :</p>


```python

L = np.linspace(0, 1000, 1000)
M = np.array([L, L, L])

```

```python

P = prod_matrices(M, M.T)
P

```

<h4 id="Loi-normale">Loi normale<a class="anchor-link" href="#Loi-normale">¶</a></h4><p>La loi normale est très utilisée en statistiques. Elle est, entre autre, la loi limite pour d'autres lois.</p>
<p>Commençons par générer des nombres suivant une loi normale mais avec différentes valeurs pour les paramètres $\mu$ et $\sigma$.</p>
<p>Exécutez la cellule suivante et observez le résultat :</p>


```python

from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline

mu = [0, 1, 2, 3]
sigma = [1, 2, 3, 4]

x_min = -16
x_max = 16
x_nb = 100

x = np.linspace(x_min, x_max, x_nb)

for i in range(len(mu)):
    y = norm.pdf(x, mu[i], sigma[i])
    plt.plot(x, y, label='mu='+str(mu[i])+', sigma='+str(sigma[i]))
plt.legend(loc='upper right')
plt.show()

```


<p>Commentez les courbes et observez l'impact des valeurs de $\sigma$.</p>
