```python

def euclidean(u, v):
    return np.linalg.norm(u-v[:-1])

```


<h4 id="Etape-2-:-Récupérer-la-liste-des-$k$-voisins-les-plus-proches">Etape 2 : Récupérer la liste des $k$ voisins les plus proches<a class="anchor-link" href="#Etape-2-:-Récupérer-la-liste-des-$k$-voisins-les-plus-proches">¶</a></h4>



<ol>
<li>Ecrire une fonction <code>voisins</code> permettant de récupérer la liste des $k$-voisins d'un sommet les plus proches dans un dataset.</li>
</ol>


```python
def voisins(u,dataset,k):
    dists = []
    for ex in dataset:
        dists.append((ex,euclidean(u,ex)))
    dists.sort(key = lambda elt:elt[1])
    _voisins = []
    for i in range(k):
        _voisins.append(dists[i][0])
    return _voisins
```


<p>Tester votre fonction. Les instructions suivantes devront produire le résultat de la cellule.</p>


```python

u = np.array([1, 2, 3])
dataset = np.array([[1, 2, 3, 1],
                   [2, 3, 4, 1],
                   [3, 4, 5, 0],
                   [4, 5, 6, 0]])
print(voisins(u, dataset, 2))

```

<li>Ecrire une fonction <code> classifier()</code> retournant la classe d'un élément $u$.</li>
</ol>


```python

def classifier(u, dataset, k):
    voisins = voisins(u,dataset,k)
    for voisin in voisins:
        if (u == voisin):
            return dataset
```
