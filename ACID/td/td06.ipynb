<h4 id="Avant-de-commencer">Avant de commencer<a class="anchor-link" href="#Avant-de-commencer">¶</a></h4><p>Ecrire une fonction <code> droite2DVd(x, vd, point) </code> permettant de calculer les ordonnées des points <code>(x, ?)</code> de la droite de vecteur directeur <code>vd</code> passant par le point <code>point</code>.
Tester votre fonction.</p>


```python
def droite2DVd(x, vd, point):
    X = 0
    Y = 1
    assert(vd[X] and vd[Y])
    coeff = vd[Y] / vd[X]
    return coeff * x + point[X] - coeff * point[Y]

x = 2
vd = [2,1]
point = [1,1]

2*2 + 1 - 2*1
droite2DVd(x,vd,point)

```

```python
from numpy.linalg import eig
import matplotlib.pyplot as plt

def CalculACP(echantillon, k):
    n = echantillon.shape[0]
    S = (n-1) * echantillon.cov()
    lambdas ,vect = eig(S)
    sorted_i = np.argsort(lambdas)[::-1]
    sorted_vec = vect[:, sorted_i]
    return sorted_vec[:, 0:k]

acp = CalculACP(data,2)
plt.scatter(acp, np.ones(len(acp)))
```
